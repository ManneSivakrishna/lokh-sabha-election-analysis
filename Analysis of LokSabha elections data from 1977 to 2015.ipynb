# Project name:Analysis of LokSabha elections data from 1977 to 2015

This database contains detailed candidate‐level data for elections to the lower houses of India’s national
and state legislatures, i.e., the Lok Sabha . The data span 1977‐2015, with each row
representing a candidate that ran for office in that state‐year.

The variables in the national election data file are:

Variable name : Storage    : type Variable contents
st_name       : str35      : State
Year          : Int        : General election year
pc_no         : Byte       : Parliamentary constituency number
pc_name       : str25      : Parliamentary constituency name
pc_type       : str3       : Parliamentary constituency reservation status
cand_name     : str70      : Candidate name
cand_sex      : str1       : Candidate sex
partyname     : str57      : Party name
partyabbre    : str10      : Party abbreviation
totvotpoll    : Long       : Votes received
electors      : Long       : Number of registered voters \n

These data can be used to calculate a number of fields, including turnout, vote shares, winning status,
effective number of parties and so forth.
[https://jovian.ai/learn/data-analysis-with-python-zero-to-pandas]

#### Q1: How many totvotpoll does each parties got from 1977 to 2014 ? ####Top 10 parties who got most of the totvotpoll from 1977 to 2014?

````python
#How many totvotpoll does each parties got from 1977 to 2014?
#Top 10 parties who got most of the totvotpoll from 1977 to 2014?
````

````python
data1=data[['partyname','partyabbre','totvotpoll','electors']]
````

````python
x=data1.groupby('partyabbre').sum()
x
````

````python
y=x.sort_values(by=['totvotpoll'],ascending=False).head(n=10)
y
````

````python
y.reset_index(inplace=True)
y
````

````python
sb.set_theme(style="whitegrid")
plt.figure(figsize=(15, 8))
sb.barplot(x="partyabbre", y="totvotpoll", data=y)
````

#### Q2: How many totvotpoll and electors from each states in 2014?

````python
#How many totvotpoll and electors from each states in 2014?
````

````python
z=data.query('year==2014')
z
````

````python
Z=z.groupby("st_name")[['partyabbre','totvotpoll','electors']].sum()
Z.reset_index(inplace=True)
Z
````

#### Q3: #Top 10 states which got most of the  electors in 2014? #Top 10 states which got most of the  totvotpoll  in 2014?

````python
ele=Z.sort_values(by=['electors'],ascending=False).head(n=10)
ele
````

````python
tot_votpoll=Z.sort_values(by=['totvotpoll'],ascending=False).head(n=10)
tot_votpoll
````

#### Q4: #Top 5 states which got least of the  electors in 2014? #Top 5 states which got least of the  totvotpoll  in 2014?

````python
ele_least=Z.sort_values(by=['electors']).head(n=5)
ele_least
````

````python
tot_votpoll_least=Z.sort_values(by=['totvotpoll']).head(5)
tot_votpoll_least
````
